Generate reports<br/>
Now, we're going to practice creating a script, named ticky_check.py, that generates two different reports from this internal ticketing system log file i.e., syslog.log. This script will create the following reports:
<br/>
<b>The ranking of errors generated by the system:</b> A list of all the error messages logged and how many times each error was found, sorted by the most common error to the least common error. This report doesn't take into account the users involved.<br/>
<b>The user usage statistics for the service:</b> A list of all users that have used the system, including how many info messages and how many error messages they've generated. This report is sorted by username.

```python
#!/usr/bin/env python3
import re
import csv
import operator
per_user={}
error={}
to_search=r": ([A-Z]*) ([\w ']*) [\[\]\d# ]*\(([\w\.]*)\)$"
with open("syslog.log", "r") as file:
  for line in file.readlines():
    result=re.search(to_search, line)
    log_type=result.group(1)
    log_message=result.group(2)
    log_user=result.group(3)
    if log_type=="ERROR":
      if log_message not in error:
        error[log_message]=1
      else:
        error[log_message]+=1
    if log_user not in per_user:
        per_user[log_user] = {"INFO": 0, "ERROR": 0}
    if log_type == "INFO":
        per_user[log_user][log_type] += 1
    elif log_type == "ERROR":
        per_user[log_user][log_type] += 1
error=sorted(error.items(), key = operator.itemgetter(1), reverse=True)
per_user=sorted(per_user.items())
with open("error_message.csv", "w") as file_err:
    writer=csv.writer(file_err)
    writer.writerow(["Error", "Count"])
    writer.writerows(error)
with open("user_statistics.csv", "w") as usr_file:
    usr_write=csv.writer(usr_file)
    usr_write.writerow(["Username", "INFO", "ERROR"])
    for item in per_user:
        user, log_type = item
        string=[user, log_type["INFO"], log_type["ERROR"]]
        usr_write.writerow(string)
```
